```python
%pip install seaborn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import zipfile
```

```python
import zipfile
import os
zip_file = 'algerian+forest+fires+dataset.zip'
if not (os.path.exists("Algerian_forest_fires_dataset_UPDATE.csv")):
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall()
    print('Zip file contents extracted to current working directory!')
```

```python
df = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header = 1)
```

```python
df.head(25)
```

```python
df.loc[:122,"Region"] = 0
df.loc[122:,"Region"] = 1
```

```python
df.head()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df = df.dropna().reset_index(drop = True)
```

```python
df.loc[100:130]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
int_columns = ['day', 'month', 'year', 'Temperature', 'RH', 'Ws' ]
```

```python
for col in int_columns:
    df[[col]] = df[[col]].astype(int)
```

```python
for col in df.columns:
    if(col not in int_columns and col!="Classes"):
        df[[col]] = df[[col]].astype(float)
```

```python
df[["Classes"]].value_counts()
```

```python
df["Classes"] = np.where(df["Classes"].str.contains("not fire"), 0,1)
```

```python
df.head()
```

```python
df[["Classes"]].value_counts()
df[["Classes"]] = df[["Classes"]].astype(int)
```

```python
if not os.path.exists("Algerian_forest_fires_dataset_CLEANED.csv"):
    df.to_csv("Algerian_forest_fires_dataset_CLEANED.csv")
```

```python
plt.style.use("seaborn")
df.hist(bins = 50,figsize = (20,15))
plt.show()
```

```python
class_labels = [0,1]
percentages = df[["Classes"]].value_counts(normalize=True)*100
plt.figure(figsize = (10,7))
plt.pie(percentages,labels =class_labels,autopct='%1.1f%%' )
plt.title("Algerian Forest Fire Statistics")
```
